```python
# Create a function that checks whether the given number is Odd or Even 
# OddEven() 
# Enter a number: 52452 
# 52452 is Even number 
```

```python
def OddEven():
    num = int(input("Enter a number:"))
    if((num%2) == 0):
        print(num,"is Even Number")
    else:
        print(num,"is Odd Number")
```

```python
# Create a function that tells elegibility of marriage for male and female according to their age limit like 21 for male and 18 for female 
# ElegiblityForMarriage.Elegible() 
# Your Gender:Male 
# Your Age:20 
# NOT ELIGIBLE 
```

```python
def ElegiblityForMarriage():
    Elegible = (input("Your Gender:"))
    Age = int(input("Your Age:"))
    if (Elegible == 'Male'):
        if Age >= 21:
            print('ELIGIBLE')
        else:
            print('NOT ELIGIBLE')
    elif (Elegible == 'Female'): 
        if Age >= 18:
            print('ELIGIBLE')
        else:
            print('NOT ELIGIBLE')
```

```python
ElegiblityForMarriage()
```

```python
def ElegiblityForMarriage():
    Elegible = (input("Your Gender:"))
    Age = int(input("Your Age:"))
    if ((Elegible == 'Male' and Age >= 21) or (Elegible == 'Female' and Age >= 18)):
        print('ELIGIBLE')
    else:
        print('NOT ELIGIBLE')
        
```

```python
#print area and perimeter of triangle using class and functions 
#triangle.triangle() 
#Height:32 
#Breadth:34 
#Area formula: (Height*Breadth)/2 
#Area of Triangle:  544.0 
#Height1:2 
#Height2:4 
#Breadth:4 
#Perimeter formula: Height1+Height2+Breadth 
#Perimeter of Triangle:  10
```

```python
Height = int(input('Height:'))
Breadth = int(input('Breadth:'))
def AreaOfTriangle():
    AreaTriangle = (Height*Breadth)/2
    print('Area of Triangle:',AreaTriangle)
```
